```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_csv('/content/cleanstudentscomplete.csv')
df
```

```python
sexo=pd.get_dummies(df[['gender']])
df1=pd.concat([df['grade'],sexo],axis=1)
df1=df1.groupby('grade').sum()
```

```python
df1.plot(kind='bar', color=['violet','orange'],figsize=(20,9),xlabel='Grado',ylabel='Numero de Alumnos')
```
